```python
# Changing target column (Categorial into numerical data)
df['species'] = df['species'].map({'setosa' : 0, 'versicolor' : 1, 'virginica' : 2})
df.head()
```

```python
# Outlier removal by IQR method
Q1 = df['sepal_width'].quantile(0.25)
Q3 = df['sepal_width'].quantile(0.75)
IQR = Q3 - Q1
```

```python
# Checking if outlier if removed or not
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df = df[(df['sepal_width'] >= lower_bound) & (df['sepal_width'] <= upper_bound)]
```

```python
# checking Boxplot for whole dataset to check whether outliers have or not
sns.boxplot(data=df)
plt.title("Iris Dataset Boxplot")
plt.show()
```

```python
# Correlation after outlier removal
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.title("Iris Dataset Correlation")
plt.show()
```

```python
sns.scatterplot(x=df['sepal_length'], y = df['sepal_width'], data = df, hue='species')
plt.title("Iris Dataset Scatter Plot - Sepal Length vs Sepal Width by Species")
plt.show()
```

```python
sns.scatterplot(x=df['petal_length'], y =df['petal_width'], data = df, hue = 'species')
plt.title("Iris Dataset Scatter Plot - petal Length vs petal Width by Species")
```
