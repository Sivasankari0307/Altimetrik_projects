```python
import pandas as pd
```

```python
# Sample dataframe
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
# Filtering
filtered_df = df[df['Age'] > 30]
filtered_df
```

```python
# Map
city_mapping = {'New York': 'NY', 'San Francisco': 'SF', 'Los Angeles': 'LA', 'Chicago': 'CH'}
df['State'] = df['City'].map(city_mapping)
df
```

```python
# Map
filter_df = df[df['Age'] > 30].copy()
filter_df
```

```python
filter_df.loc[:, 'Age'] = filter_df['Age'].apply(lambda x: x + 1)
filter_df
```

```python
# Reduce
filter_df_avg = filter_df['Age'].mean()
filter_df_avg
```
