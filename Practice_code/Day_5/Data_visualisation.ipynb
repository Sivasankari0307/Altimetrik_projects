```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# example for data visualisation
```

```python
# Line chart
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]
attendance = [90, 85, 95, 92, 88]
plt.plot(days, attendance, marker='o', color='blue')
plt.title("Attendance Over Days")
plt.xlabel("Days")
plt.ylabel("Attendance")
plt.grid(False)
plt.show()
```

```python
# Bar chart
categories = ["A", "B", "C", "D"]
values = [10, 15, 8, 12]
plt.bar(categories, values, color=['red', 'green', 'blue', 'orange'])
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
# Pie chart
subjects = ['Tamil','English','Maths','Science','Social']
marks = [95,65,99,85,80]
plt.pie(marks,labels=subjects,autopct='%1.1f%%', startangle = 90)
plt.title('Marks Distribution')
plt.legend(bbox_to_anchor=(1.2, 0.5), loc=2)
plt.show()
```

```python
# scatter chart
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
plt.scatter(x, y, color='red', marker='o')
plt.title("Scatter Plot Example")
plt.xlabel("Marks")
plt.ylabel("Student_id")
plt.grid(True)
plt.show()
```

```python
# Histogram chart ( shows the frequency distribution of a dataset )
data = [23,45,67,45,34,67,34,56,12]

plt.hist(data, bins=5, color='green', edgecolor='black')
plt.title("Histogram Example")
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```
