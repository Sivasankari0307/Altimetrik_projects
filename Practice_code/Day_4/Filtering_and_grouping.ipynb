```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("sales.csv")
df
```

```python
df.groupby("Product")["Unit_Price"].sum()
```

```python
# Multiple aggregration
df.groupby("Product")["Unit_Price"].agg(["sum","mean"])
```

```python
df.groupby("Product")["Unit_Price"].mean()
```

```python
# Filtering
df.groupby("Product").filter(lambda x: x["Unit_Price"].mean() > 10)
```

```python
# Grouping
df.groupby("Product").groups
```
