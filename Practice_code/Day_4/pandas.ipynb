```python
# Use case
employee = {
    "emp_id" : [101,102,103, 104],
    "emp_name" : ["Siva", "Iniya", "priya","Sankari"],
    "dept" : ["Data", "Java", "Java","Data"],
    "salary" : [25000, 30000, 50000, 40000]}

df = pd.DataFrame(employee)
df
```

```python
df[df["salary"]>30000]
```
